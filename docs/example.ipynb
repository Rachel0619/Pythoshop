# Example usage

To use `pythoshop` in a project:

## Imports

```python
from pythoshop.adjust_brightness import adjust_brightness
from pythoshop.resize_image import resize_image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## adjust_brightness

This function changes the brightness of an input image by a specified factor. The factor can be positive to increase brightness or negative to decrease it. The brightness change is applied uniformly to all pixels of the image. The brightness_factor can be int or float, corresponding to abosulte change or percentage change

### Before change

```python
img = mpimg.imread("../tests/test_img_1.png")
plt.imshow(img)
plt.axis('off')  # Turn off axis numbers and labels
plt.show()
```

### After change

```python
adjust_brightness("../tests/test_img_1.png", 0.5)
```

```python
img = mpimg.imread("../tests/test_img_1_brightness_changed.png")
plt.imshow(img)
plt.axis('off')  # Turn off axis numbers and labels
plt.show()
```

## resize_image
 This section demonstrates three use cases of the resize_image function: cropping, maintaining aspect ratio, and adding borders. It also displays the original and resized images side by side for comparison.

```python
# Define example image path
input_path = "../tests/mds_logo.png"

# Example 1: Resize image with cropping
resize_image(input_path, height=150, width=150, method='crop')

# Example 2: Resize image maintaining aspect ratio
resize_image(input_path, height=180, width=180, method='maintain_aspect_ratio')

# Example 3: Resize image with added borders
resize_image(input_path, height=50, width=50, method='add_borders')

# Display the original and resized images
original_img = plt.imread(input_path)
resized_img = plt.imread(f"{input_path[:-4]}_res_img.png")

# Plot original image
print("The Original Image")
plt.subplot(1, 2, 1)
plt.imshow(original_img)
plt.title("Original Image")
plt.axis('off')

```
